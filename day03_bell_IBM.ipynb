```python
# Day 3 - Bell State Circuit - IBM Qiskit
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_state_city

# Build Bell State (H + CNOT)
qc = QuantumCircuit(2)
qc.h(0)      # H gate on qubit 0
qc.cx(0, 1)  # CNOT gate - qubit 0 controls qubit 1

# Draw the circuit
print(qc.draw())

# Get the state
sv = Statevector(qc)
print("\nState:", sv)
print("\nResult: 50% |00> and 50% |11> - Bell State confirmed!")
```

```python
# Day 3 - Bell State Circuit - Google Cirq
import cirq

# Create 2 qubits
q0, q1 = cirq.LineQubit.range(2)

# Build Bell State
circuit = cirq.Circuit([
    cirq.H(q0),           # H gate on qubit 0
    cirq.CNOT(q0, q1),    # CNOT gate
    cirq.measure(q0, q1, key='result')
])

# Draw the circuit
print(circuit)

# Simulate
simulator = cirq.Simulator()
result = simulator.run(circuit, repetitions=100)
print("\nResults:", result)
print("\nResult: 50% |00> and 50% |11> - Bell State confirmed!")
```

```python
# Day 3 - Bell State Circuit - Amazon Braket
from braket.circuits import Circuit
from braket.devices import LocalSimulator

# Build Bell State
circuit = Circuit()
circuit.h(0)        # H gate on qubit 0
circuit.cnot(0, 1)  # CNOT gate

# Draw the circuit
print(circuit)

# Simulate locally
device = LocalSimulator()
result = device.run(circuit, shots=100).result()
counts = result.measurement_counts
print("\nMeasurement counts:", counts)
print("\nResult: 50% |00> and 50% |11> - Bell State confirmed!")
```

```python
# Day 3 - Bell State Circuit - NVIDIA CUDA-Q
import cudaq

# Build Bell State
@cudaq.kernel
def bell_state():
    qubits = cudaq.qvector(2)
    h(qubits[0])        # H gate on qubit 0
    cx(qubits[0], qubits[1])  # CNOT gate
    mz(qubits)          # Measure all

# Draw the circuit
print(cudaq.draw(bell_state))

# Simulate
result = cudaq.sample(bell_state, shots_count=100)
print("Measurement counts:", result)
print("\nResult: 50% |00> and 50% |11> - Bell State confirmed!")
```
